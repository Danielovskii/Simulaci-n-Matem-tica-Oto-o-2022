**Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.**

$f(x)=1260+117x-9x^2$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

sp.var("x")

def f(x) :
    return 1260 + 117*x - 9*x**2

df = sp.diff(f(x), x)
ddf = sp.diff(df, x)

pc1 = sp.solve(df, x)[0]

fn = sp.lambdify(x, f(x), "numpy" )

xn = np.linspace(-20, 20, 100)
yn = fn(xn)

coord_y = f(pc1)

segunda_derivada = ddf.subs(x, pc1)
```

```python
plt.figure()
plt.plot(xn, yn)
plt.grid()

plt.plot(pc1, coord_y, "*r", ms = 10)
```

```python
segunda_derivada
```

Por lo tanto el punto crítico es un máximo

$f(x)=5+x+\frac{4}{x}$, para $x>0$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

sp.var("x")

def f(x):
    return x+5+(4/x)

df = sp.diff(f(x), x)
ddf = sp.diff(df, x)

pc = sp.solve(df, x)[0]

xn = np.linspace(0.5, 20, 200)
yn = f(xn)

plt.figure()
plt.plot(xn, yn)
plt.grid()

a = pc
if a >= 0:
    pc1 = a
    coord_y = f(pc1)
    plt.plot(pc1, coord_y, "*r", ms = 10)
    ddfp = sp.diff(df, x)
    segunda_derivada = ddfp.subs(x, pc1)


```

```python
segunda_derivada
```

Por lo tanto el punto crítico es un máximo

$f(x)=6-2x+\sqrt{18-2x^2}$, para $-3\leq x\leq 3$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

sp.var("x")

def f(x) : 
    return 6 - (2*x) + ((18-(2*(x**2)))**1/2)

df = sp.diff(f(x), x)
ddf = sp.diff(df, x)

pc = sp.solve(df, x)[0]

xn = np.linspace(-10, 10, 200)
yn = f(xn)

plt.figure()
plt.plot(xn, yn)
plt.grid()

coord_y = round(f(pc))

segunda_derivada = ddf.subs(x, pc)

plt.plot(pc, coord_y, "*r", ms = 10)
```

por lo tanto el punto crítico es un máximo, y se encuentra en el intervalo $-3\leq x\leq 3$

$f(x)=\frac{1}{4}x^4+\frac{1}{3}x^3-x^2$ en el intervalo $[-3,1.7]$

```python
import sympy as sp
import numpy as np
import matplotlib as plt

sp.var("x")

plt.figure()

def f(x) : 
    return (1/4*(x**4)) + (1/3*(x**3)) - x**2

df = sp.diff(f(x), x)
ddf = sp.diff(df, x)

pc = sp.solve(df, x)

xn = np.linspace(-10, 10, 200)
yn = f(xn)

plt.plot(xn, yn)
plt.grid()

a = pc
if a >= 3 and a <= 1.7:
    pc1 = int(a)
    coord_y = round(f(pc1), 3)
    plt.plot(pc1, coord_y, "*r", ms = 10)
    ddf = sp.diff(df, x)
    segunda_derivada = ddf.subs(x, pc1)
```
