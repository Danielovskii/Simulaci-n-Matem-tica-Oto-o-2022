# Actividad - Ecuaciones diferenciales

```python
import numpy as np
from scipy.integrate import solve_ivp, odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

1. Resolver numéricamente con `solve_ivp` la siguiente ecuación diferencial:
$$\frac{dx}{dt}=\frac{x}{t+1}$$

```python
def funcion(x, t) :
    return x/(t + 1)

tspan = [0, 50]
x0 = np.array([2])

x = solve_ivp(funcion, tspan, x0, max_step = 0.05)
```

Grafique con condición inicial $x_0 = 2$

```python
plt.figure()
plt.plot(x.t, x.y[0,:])
```

2. Encuentre la solución numérica para un oscilador armónico simple:

$$
\frac{d^2 x}{dt^2} +  \omega_{0}^2 x = 0.
$$
Donde $x_0=0.5$, $\dot{x}_0=0.1$ y $\omega_0=3$.

2.1 Defina una función, cuyos argumentos de entrada sean:


1. el tiempo $t$,
2. el arreglo $[x,\dot{x}]$.

Y la salida:

1. el arreglo $[\dot{x},\ddot{x}]$.

  Considerando el argumento de la frecuencia natural $\omega_0$ debe ser una variable global.

```python
w0 = 3
x0 = 0.5
v0 = 0.1

def dX(t, X) :
    return np.array([X[1], -w0**2*X[0]])

tspan1 = [0, 10] # de 0 a 10 segundos
X = np.array([0.5, 0.1])

x1 = solve_ivp(dX, tspan1, X)
```

Grafique

```python
plt.figure()
plt.plot(x1.t, x1.y[0,:])
plt.grid()
```

```python
tspan5 = [0, 50] # de cero a 50 segundos
X = np.array([0.5, 0.1])

x5 = solve_ivp(dX, tspan5, X)

plt.figure()
plt.plot(x5.t, x5.y[0,:])
plt.grid()
```

Asigne a los parámetros de la solución analítica $C_1=x_0$ y $C_2=\frac{\dot{x}_0}{\omega_0}$.

```python
def posicion(t, w0, x0, v0) :
    return ( x0*np.cos(w0*t)) + ((v0/w0)*np.sin(w0*t) )
```

```python
t = np.linspace(0, 10, 1000) # de 0 a 10 segundos
plt.figure()
plt.plot(t, posicion(t, w0, x0, v0), c = 'r')
plt.grid()
```

```python
t2 = np.linspace(0, 50, 1000) # de 0 a 50 segundos
plt.figure()
plt.plot(t2, posicion(t2, w0, x0, v0), c = 'r')
plt.grid()
```
